## Formatting Checks

These assertions catch the previous output problems: duplicated first-line labels, raw `Dict` example formatting, and inconsistent `required; default=...` schema entries.

```julia
help_text = Dict(name(T) => sprint(element_help, T) for T in registered_element_specs())

for (kind, text) in help_text
    @assert occursin("Element kind: :$(kind)", text)
    @assert !occursin("Dict{Symbol", text)
    @assert !occursin("required; default=", text)
    @assert occursin("Construction:", text)
    @assert occursin("Parameters:", text)
    @assert occursin("Example:", text)
    @assert occursin("Related queries:", text)
end

println("element_help formatting checks passed for ", length(help_text), " registered specs")
```

## Friendly Constructor Query Checks

Users should be able to query familiar names such as `LorentzBoostSpec`, not only the raw `ElementSpec{kind}` type.

```julia
friendly_specs = (
    CrabDispersionSpec,
    MomentumDispersionSpec,
    XYCouplingSpec,
    ThinCrabCavitySpec,
    LorentzBoostSpec,
    RevLorentzBoostSpec,
)

for F in friendly_specs
    println("\n", repeat("-", 80))
    element_help(F)
    @assert !isempty(construction_help(F))
    @assert example_spec(F) isa ElementSpec
    @assert !isempty(keys(parameter_schema(F)))
end

println("friendly constructor query checks passed")
```

## Raw And Friendly Metadata Agreement

The raw flexible spec type and the friendly constructor name should point to the same schema, example kind, and construction help.

```julia
raw_friendly_pairs = (
    (ElementSpec{:crab_dispersion}, CrabDispersionSpec),
    (ElementSpec{:momentum_dispersion}, MomentumDispersionSpec),
    (ElementSpec{:xy_coupling}, XYCouplingSpec),
    (ElementSpec{:thin_crab_cavity}, ThinCrabCavitySpec),
    (ElementSpec{:lorentz_boost}, LorentzBoostSpec),
    (ElementSpec{:rev_lorentz_boost}, RevLorentzBoostSpec),
)

for (raw, friendly) in raw_friendly_pairs
    @assert parameter_schema(raw) == parameter_schema(friendly)
    @assert construction_help(raw) == construction_help(friendly)
    @assert kind(example_spec(raw)) == kind(example_spec(friendly))
end

println("raw and friendly metadata agreement checks passed")
```

## Existing Spec Object Query

`element_help` also accepts an already constructed spec object.

```julia
spec = ThinCrabCavitySpec{2}(
    1.0e8;
    strengthX = (1.0, 0.5),
    strengthY = (0.25, 0.125),
    phase = (0.0, 0.1),
    aperture = (1.0e-3, 2.0e-3),
)

element_help(spec)
@assert kind(spec) == :thin_crab_cavity
@assert hasparam(spec, :aperture)
@assert getparam(spec, :aperture) == (1.0e-3, 2.0e-3)
```
